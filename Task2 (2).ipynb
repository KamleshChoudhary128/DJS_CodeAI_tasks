```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("deliveries_updated_mens_ipl_upto_2024 - deliveries_updated_mens_ipl_upto_2024.csv")
data.head()
data.describe()
```

```python
data.fillna(0,inplace=True)
data.info()
```

```python
plt.figure(figsize=(10,10))
runs=data.groupby('batting_team')['batsman_runs'].sum().sort_values(ascending=True).head(8)
plt.bar(runs.index,runs.values,color='gold')
plt.xlabel('Total Runs')
plt.ylabel('Teams')
plt.title('According to total runs top 8 teams',fontweight='bold')
plt.xticks(rotation=30, ha='right')
plt.tight_layout
plt.show()
```

```python
plt.figure(figsize=(14,8))
runs1=data.groupby('over')['batsman_runs'].mean()
plt.plot(runs1.index,runs1.values,markersize=8,marker='.',linewidth=2,color='darkorange',label='Avg runs')
plt.title('Average runs per Over',fontweight='bold',fontsize=15)
plt.xlabel('Overs')
plt.ylabel('Average runs per ball')
plt.legend()
plt.show()
```

```python
data['year']=pd.to_datetime(data['date']).dt.year
plt.figure(figsize=(14,8))
runs2=data.groupby('year')['batsman_runs'].sum()
plt.plot(runs2.index,runs2.values,marker='.',markersize=8,linewidth=3,color='teal',label='Runs per year')
plt.title('runs scored per year',fontsize=15,fontweight='bold')
plt.ylabel('Total runs')
plt.xlabel('Years')
plt.legend()
plt.show()
```
